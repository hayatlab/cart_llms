````python
import os
import pandas as pd
import time
import re
import json
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from tqdm import tqdm
import logging
from datetime import datetime
import glob
import scipy
import math
# Set up logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler("weight_reanalysis.log"),
        logging.StreamHandler()
    ]
)
logger = logging.getLogger("weight_reanalysis")

def load_existing_results_from_raw(base_path="/data/ep924610/project_nb/paper_code/bulk_review"):
    """Load existing results from individual JSON files in the raw directories"""
    openai_results = []
    claude_results = []
    gemini_results = []
    
    # Different paths for different models
    llm_raw_dir = f"{base_path}/llm/raw"
    gemini_raw_dir = f"{base_path}/results/raw"
    
    try:
        # Load OpenAI results
        openai_files = glob.glob(f"{llm_raw_dir}/openai_run_*.json")
        logger.info(f"Found {len(openai_files)} OpenAI result files")
        
        for file in tqdm(openai_files, desc="Loading OpenAI results"):
            try:
                with open(file, "r") as f:
                    result = json.load(f)
                    openai_results.append(result)
            except Exception as e:
                logger.warning(f"Error loading {file}: {str(e)}")
        
        # Load Claude results
        claude_files = glob.glob(f"{llm_raw_dir}/claude_run_*.json")
        logger.info(f"Found {len(claude_files)} Claude result files")
        
        for file in tqdm(claude_files, desc="Loading Claude results"):
            try:
                with open(file, "r") as f:
                    result = json.load(f)
                    claude_results.append(result)
            except Exception as e:
                logger.warning(f"Error loading {file}: {str(e)}")
        
        # Load Gemini results from different directory
        gemini_files = glob.glob(f"{gemini_raw_dir}/gemini_run_*.json")
        logger.info(f"Found {len(gemini_files)} Gemini result files")
        
        for file in tqdm(gemini_files, desc="Loading Gemini results"):
            try:
                with open(file, "r") as f:
                    result = json.load(f)
                    gemini_results.append(result)
            except Exception as e:
                logger.warning(f"Error loading {file}: {str(e)}")
        
        # Sort results by run_id for consistency
        for results in [openai_results, claude_results, gemini_results]:
            results.sort(key=lambda x: x.get("run_id", 0))
        
        logger.info(f"Loaded {len(openai_results)} OpenAI, {len(claude_results)} Claude, and {len(gemini_results)} Gemini results")
            
    except Exception as e:
        logger.error(f"Error loading existing results: {str(e)}")
    
    return openai_results, claude_results, gemini_results

def extract_weights(response_text):
    """Extract weight recommendations from response text with improved pattern matching"""
    # Try the exact formatted version first (standard pattern)
    pattern1 = r"RECOMMENDED WEIGHTS:\s*Clinical trials \((\d+\.\d+)\),\s*HPA \((\d+\.\d+)\),\s*Heatmap \((\d+\.\d+)\),\s*Vital \((\d+\.\d+)\),\s*Non-vital \((\d+\.\d+)\),\s*Protein vital \((\d+\.\d+)\),\s*Protein non-vital \((\d+\.\d+)\)"
    match = re.search(pattern1, response_text, re.IGNORECASE)
    
    if match:
        return {
            "Clinical_trials": float(match.group(1)),
            "HPA": float(match.group(2)),
            "Heatmap": float(match.group(3)),
            "Vital": float(match.group(4)),
            "Non_vital": float(match.group(5)),
            "Protein_vital": float(match.group(6)),
            "Protein_non_vital": float(match.group(7))
        }
    
    # Try alternative format with bold/markdown formatting
    pattern2 = r"\*\*RECOMMENDED WEIGHTS:?\*\*:?\s*Clinical trials \((\d+\.\d+)\),\s*HPA \((\d+\.\d+)\),\s*Heatmap \((\d+\.\d+)\),\s*Vital \((\d+\.\d+)\),\s*Non-vital \((\d+\.\d+)\),\s*Protein vital \((\d+\.\d+)\),\s*Protein non-vital \((\d+\.\d+)\)"
    match = re.search(pattern2, response_text)
    
    if match:
        return {
            "Clinical_trials": float(match.group(1)),
            "HPA": float(match.group(2)),
            "Heatmap": float(match.group(3)),
            "Vital": float(match.group(4)),
            "Non_vital": float(match.group(5)),
            "Protein_vital": float(match.group(6)),
            "Protein_non_vital": float(match.group(7))
        }
    
    # Try a more flexible pattern for each individual weight
    weights = {}
    patterns = {
        "Clinical_trials": r"Clinical trials\s*\(?(\d+\.?\d*)\)?",
        "HPA": r"HPA\s*\(?(\d+\.?\d*)\)?",
        "Heatmap": r"Heatmap\s*\(?(\d+\.?\d*)\)?",
        "Vital": r"(?<!\w)Vital\s*\(?(\d+\.?\d*)\)?",  # negative lookbehind to avoid matching "Protein vital"
        "Non_vital": r"Non-vital\s*\(?(\d+\.?\d*)\)?",
        "Protein_vital": r"Protein vital\s*\(?(\d+\.?\d*)\)?",
        "Protein_non_vital": r"Protein non-vital\s*\(?(\d+\.?\d*)\)?"
    }
    
    # Look for individual weights
    for key, pattern in patterns.items():
        match = re.search(pattern, response_text)
        if match:
            try:
                weights[key] = float(match.group(1))
            except (ValueError, TypeError):
                # Skip if we can't convert to float
                continue
    
    # Return weights if we found at least 4 features
    if len(weights) >= 4:
        return weights
    
    # If all else fails, try a line-by-line approach to find weight numbers
    if len(weights) < 4:
        # Look for lines like "1. Heatmap (5.0):" or "1. **Heatmap (5.0)**:"
        for line in response_text.split('\n'):
            for key in patterns.keys():
                # Convert key to a readable format
                readable_key = key.replace('_', ' ')
                if readable_key in line:
                    # Extract numbers from the line
                    numbers = re.findall(r'(\d+\.\d+)', line)
                    if numbers:
                        try:
                            weights[key] = float(numbers[0])
                        except (ValueError, TypeError, IndexError):
                            continue
        
        # Return weights if we found at least 4 features
        if len(weights) >= 4:
            return weights
    
    return None

def reanalyze_all_models(base_path="/data/ep924610/project_nb/paper_code/bulk_review"):
    """Re-analyze results from all three models using consistent extraction"""
    # Load existing results from raw files
    openai_results, claude_results, gemini_results = load_existing_results_from_raw(base_path)
    
    weight_features = ["Clinical_trials", "HPA", "Heatmap", "Vital", 
                      "Non_vital", "Protein_vital", "Protein_non_vital"]
    
    # Process results for each model
    def process_results(results, model_name):
        model_weights = []
        success_count = 0
        failed_count = 0
        
        for result in tqdm(results, desc=f"Processing {model_name}"):
            try:
                weights = extract_weights(result["response"])
                if weights:
                    weights["model"] = model_name
                    weights["run_id"] = result["run_id"]
                    weights["temperature"] = result.get("temperature", np.nan)
                    model_weights.append(weights)
                    success_count += 1
                else:
                    failed_count += 1
                    logger.debug(f"Could not extract weights from {model_name} run {result.get('run_id', 'unknown')}")
            except Exception as e:
                logger.error(f"Error processing {model_name} result {result.get('run_id', 'unknown')}: {str(e)}")
                failed_count += 1
        
        logger.info(f"Extracted weights from {success_count}/{len(results)} {model_name} responses. Failed: {failed_count}")
        
        # Quick check of some sample failures (log first 3 failed responses)
        if failed_count > 0:
            fail_count = 0
            for result in results:
                if extract_weights(result["response"]) is None:
                    logger.debug(f"Sample failed response from {model_name} run {result.get('run_id', 'unknown')}: {result['response'][:200]}...")
                    fail_count += 1
                    if fail_count >= 3:
                        break
        
        return model_weights
    
    # Extract weights from each model
    openai_weights = process_results(openai_results, "GPT-4o")
    claude_weights = process_results(claude_results, "Claude-3.7")
    gemini_weights = process_results(gemini_results, "Gemini-2.5-Pro")
    
    # Combine all weights into a DataFrame
    all_weights = pd.DataFrame(openai_weights + claude_weights + gemini_weights)
    
    # Output directory for reanalyzed data
    output_dir = f"{base_path}/reanalysis"
    os.makedirs(output_dir, exist_ok=True)
    
    # Save extracted weights
    all_weights.to_csv(f"{output_dir}/all_models_weights_reanalyzed.csv", index=False)
    logger.info(f"Saved extracted weights from all models. Total: {len(all_weights)} rows")
    
    # Create summary statistics
    summary = all_weights.groupby("model")[weight_features].agg(['mean', 'median', 'std', 'min', 'max'])
    summary.to_csv(f"{output_dir}/all_models_summary_stats_reanalyzed.csv")
    
    # Count most common weight combinations by model
    def get_weight_combo(row):
        return tuple(round(row[feature], 1) for feature in weight_features)
    
    all_weights["weight_combo"] = all_weights.apply(get_weight_combo, axis=1)
    combo_counts = all_weights.groupby(["model", "weight_combo"]).size().reset_index(name="count")
    combo_counts = combo_counts.sort_values(["model", "count"], ascending=[True, False])
    
    # Save top combinations
    top_combos = pd.DataFrame()
    for model in all_weights["model"].unique():
        model_combos = combo_counts[combo_counts["model"] == model].head(5)
        model_combos["weight_combo_str"] = model_combos["weight_combo"].apply(
            lambda x: ", ".join([f"{feat}={val}" for feat, val in zip(weight_features, x)])
        )
        top_combos = pd.concat([top_combos, model_combos])
    
    top_combos.to_csv(f"{output_dir}/all_models_top_combinations_reanalyzed.csv", index=False)
    
    # Save consolidated results (for future use)
    with open(f"{output_dir}/openai_all_results_reanalyzed.json", "w") as f:
        json.dump(openai_results, f, indent=2)
        
    with open(f"{output_dir}/claude_all_results_reanalyzed.json", "w") as f:
        json.dump(claude_results, f, indent=2)
        
    with open(f"{output_dir}/gemini_all_results_reanalyzed.json", "w") as f:
        json.dump(gemini_results, f, indent=2)
    
    logger.info(f"Analysis complete for all models")
    return all_weights, summary, top_combos

def create_three_model_visualizations(all_weights, summary, base_path="/data/ep924610/project_nb/paper_code/bulk_review"):
    """Create visualizations comparing all three models"""
    import scipy.stats as stats
    
    vis_path = f"{base_path}/reanalysis/visualizations"
    os.makedirs(vis_path, exist_ok=True)
    
    # Set color palette for models - using a darker color scheme
    model_colors = {
        "GPT-4o": "#1d6657",       # Dark green for OpenAI
        "Claude-3.7": "#b34700",   # Dark orange for Anthropic
        "Gemini-2.5-Pro": "#0b5394" # Dark blue for Google
    }
    
    # Feature names for plotting with original database column names
    weight_features = ["Clinical_trials", "HPA", "Heatmap", "Vital", 
                      "Non_vital", "Protein_vital", "Protein_non_vital"]
    
    # Mapping of database column names to display names with updated specificity naming
    feature_display_names = {
        "Clinical_trials": "Translatability",
        "HPA": "Subcellular Location",
        "Heatmap": "Cancer Composition",
        "Vital": "Specificity - Vital - Bulk RNA",
        "Non_vital": "Specificity - Non Vital - Bulk RNA",
        "Protein_vital": "Specificity - Vital - Protein",
        "Protein_non_vital": "Specificity - Non Vital - Protein"
    }
    
    # Convert weights to long format for violin plots
    weights_long = all_weights.melt(id_vars=["model", "run_id"], 
                                  value_vars=weight_features,
                                  var_name="feature", value_name="value")
    
    # 1. Violin plots with modified styling, individual dots, and both scaling options
    logger.info("Creating distribution violin plots...")
    
    def violin_plots_simple(weights_long, weight_features, model_colors, feature_display_names):
        import matplotlib.pyplot as plt
        import seaborn as sns
        import math

        fig = plt.figure(figsize=(12, 18))
        row_count = math.ceil(len(weight_features) / 2)
        for i, feature in enumerate(weight_features):
            ax = plt.subplot(row_count, 2, i+1)
            sns.violinplot(
                x="model",
                y="value",
                data=weights_long[weights_long["feature"] == feature],
                inner="box",
                palette=model_colors,
                order=["GPT-4o", "Claude-3.7", "Gemini-2.5-Pro"]
            )
            sns.stripplot(
                x="model",
                y="value",
                data=weights_long[weights_long["feature"] == feature],
                size=4, color='black', alpha=0.5, jitter=0.25, dodge=True,
                order=["GPT-4o", "Claude-3.7", "Gemini-2.5-Pro"]
            )
            ax.set_title(feature_display_names[feature], fontsize=16)
            ax.set_ylabel('Weight', fontsize=14)
            ax.set_xlabel('Reasoning Model', fontsize=14)
        plt.tight_layout()
        return fig
    
    # Create violin plots with the updated function
    # ... inside create_three_model_visualizations
    violin_fig = violin_plots_simple(weights_long, weight_features, model_colors, feature_display_names)
    violin_fig.savefig(f"{vis_path}/three_model_weight_distributions.svg", dpi=300)

    # 2. Heatmap of average weights by model with inverted axes and horizontal text
    logger.info("Creating average weights heatmap...")
    model_order = ["GPT-4o", "Claude-3.7", "Gemini-2.5-Pro"]
    avg_weights = all_weights.groupby("model")[weight_features].mean().reindex(model_order)
    
    # Create a copy with renamed columns for display
    avg_weights_display = avg_weights.copy()
    avg_weights_display.columns = [feature_display_names[col] for col in avg_weights.columns]
    
    plt.figure(figsize=(12, 10))
    # Transpose the data to invert the axes (reasoning models on y-axis)
    sns.heatmap(
        avg_weights_display.T, 
        annot=True,
        cmap="YlGnBu",
        fmt=".2f",
        linewidths=0.5,
        annot_kws={"size": 18}  # Make annotation numbers larger
    )
    plt.ylabel("Feature", fontsize=18)
    plt.xlabel("Reasoning Model", fontsize=18)
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    plt.title("Average Weights by Model", fontsize=18)
    plt.legend(loc='upper right', bbox_to_anchor=(0.2, 0.1), fontsize=18)
    plt.tight_layout()
    plt.savefig(f"{vis_path}/three_model_average_weights_heatmap.svg", dpi=300)
    
    # 3. Radar chart with modified styling
    logger.info("Creating radar chart...")
    def radar_chart(avg_weights, weight_features, model_colors, feature_display_names):
        # Number of variables
        N = len(weight_features)
        
        # What will be the angle of each axis in the plot
        angles = [n / float(N) * 2 * np.pi for n in range(N)]
        angles += angles[:1]
        
        # Initialize the plot
        fig = plt.figure(figsize=(12, 12))
        ax = fig.add_subplot(111, polar=True)
        
        # Calculate maximum value to set appropriate yticks (rounded up to nearest integer)
        max_val = np.ceil(avg_weights.values.max())
        
        # Draw one axis per variable and add labels with display names
        plt.xticks(angles[:-1], [feature_display_names[f] for f in weight_features], size=14)
        
        # Draw ylabels as integer values
        ax.set_rlabel_position(0)
        yticks = np.arange(1, max_val+1, dtype=int)
        plt.yticks(yticks, [str(int(y)) for y in yticks], color="grey", size=12)
        plt.ylim(0, max_val)
        
        # Plot each model with custom colors
        for model in avg_weights.index:
            values = avg_weights.loc[model].values.tolist()
            values += values[:1]
            ax.plot(angles, values, linewidth=3, linestyle='solid', label=model, color=model_colors.get(model, "gray"))
            ax.fill(angles, values, alpha=0.2, color=model_colors.get(model, "gray"))
        
        # Add legend
        plt.legend(loc='upper right', bbox_to_anchor=(0.2, 0.1), fontsize=14)
        
        # Add title with more space
        plt.title("Weight Comparison Across Models", size=20, y=1.1)
        
        # Crucial step: expand the figure size to make room for labels
        plt.tight_layout(pad=5.0)
        
        return fig
    
    radar_fig = radar_chart(avg_weights, weight_features, model_colors, feature_display_names)
    radar_fig.savefig(f"{vis_path}/three_model_radar_chart.svg", dpi=300, bbox_inches='tight')
    
    # 4. Perform comprehensive ANOVA tests for the report
    logger.info("Performing ANOVA tests across all features...")
    
    def perform_anova_tests(weights_long, weight_features):
        # Create a DataFrame to store ANOVA results
        anova_results = pd.DataFrame(columns=['Feature', 'F-statistic', 'p-value', 'Significant'])
        
        # Perform ANOVA for each feature
        for feature in weight_features:
            # Get data for this feature
            feature_data = weights_long[weights_long['feature'] == feature]
            
            # Group data by model
            model_groups = [group['value'].values for name, group in feature_data.groupby('model')]
            
            # Perform one-way ANOVA
            f_stat, p_val = stats.f_oneway(*model_groups)
            
            # Determine significance
            significant = 'Yes' if p_val < 0.05 else 'No'
            
            # Add to results DataFrame
            anova_results = anova_results._append({
                'Feature': feature,
                'F-statistic': f_stat,
                'p-value': p_val,
                'Significant': significant
            }, ignore_index=True)
        
        # Sort by significance and p-value
        anova_results = anova_results.sort_values(by=['Significant', 'p-value'], ascending=[False, True])
        
        return anova_results
    
    # Run ANOVA tests and save results
    anova_results = perform_anova_tests(weights_long, weight_features)
    anova_results.to_csv(f"{vis_path}/anova_test_results.csv", index=False)
    
    # Create a visual summary of ANOVA results
    plt.figure(figsize=(12, 8))
    
    # Create a bar plot of -log10(p-value) for easier visualization of significance
    anova_results['neg_log_p'] = -np.log10(anova_results['p-value'])
    
    # Sort by p-value for the plot
    anova_sorted = anova_results.sort_values('p-value')
    
    # Create bar plot
    bars = plt.bar(
        [feature_display_names[f] for f in anova_sorted['Feature']], 
        anova_sorted['neg_log_p'],
        color=[('darkred' if p < 0.05 else 'darkblue') for p in anova_sorted['p-value']]
    )
    
    # Add significance threshold line at -log10(0.05)
    plt.axhline(y=-np.log10(0.05), color='red', linestyle='--', alpha=0.7)
    plt.text(0, -np.log10(0.05)+0.1, 'p=0.05', color='red')
    
    plt.title('ANOVA Significance: Model Differences in Feature Weights', fontsize=16)
    plt.ylabel('-log10(p-value)', fontsize=14)
    plt.xlabel('Feature', fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    
    # Add p-value annotations on top of bars
    for i, bar in enumerate(bars):
        p_value = anova_sorted['p-value'].iloc[i]
        if p_value < 0.001:
            plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1, 
                    'p<0.001', ha='center', va='bottom', rotation=0, fontsize=9)
        else:
            plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1, 
                    f'p={p_value:.3f}', ha='center', va='bottom', rotation=0, fontsize=9)
    
    plt.savefig(f"{vis_path}/anova_significance_plot.svg", dpi=300, bbox_inches='tight')
    
    logger.info(f"Visualizations complete. Saved to {vis_path}")
def generate_comparative_report(all_weights, summary, top_combos, base_path="/data/ep924610/project_nb/paper_code/bulk_review"):
    """Generate a comprehensive report comparing all three models"""
    report_path = f"{base_path}/reanalysis"
    os.makedirs(report_path, exist_ok=True)
    
    logger.info("Generating comparative report...")
    
    # Mapping of database column names to display names
    feature_display_names = {
        "Clinical_trials": "Translatability",
        "HPA": "Subcellular Location",
        "Heatmap": "Cancer Composition",
        "Vital": "Vital Organ Specificity Bulk",
        "Non_vital": "Non-Vital Organ Specificity Bulk",
        "Protein_vital": "Vital Organ Specificity Protein",
        "Protein_non_vital": "Non-Vital Organ Specificity Protein"
    }
    
    report = []
    report.append("# CAR T Target Weight Recommendation - Three Model Comparison")
    report.append(f"Analysis Date: {datetime.now().strftime('%Y-%m-%d')}")
    
    # Count of samples by model
    model_counts = all_weights.groupby("model").size()
    report.append("\n## Model Samples")
    report.append(f"- GPT-4o: {model_counts.get('GPT-4o', 0)} samples")
    report.append(f"- Claude-3.7: {model_counts.get('Claude-3.7', 0)} samples")
    report.append(f"- Gemini-2.5-Pro: {model_counts.get('Gemini-2.5-Pro', 0)} samples")
    
    # Weight features list
    weight_features = ["Clinical_trials", "HPA", "Heatmap", "Vital", 
                      "Non_vital", "Protein_vital", "Protein_non_vital"]
    
    # Add mean weights table
    report.append("\n## Average Weights by Model")
    means = all_weights.groupby("model")[weight_features].mean()
    
    # Create a copy with renamed columns for display in the report
    means_display = means.copy()
    means_display.columns = [feature_display_names[col] for col in means.columns]
    
    report.append("```")
    report.append(means_display.to_string())
    report.append("```")
    
    # Add coefficient of variation to gauge consistency
    report.append("\n## Model Consistency (Coefficient of Variation)")
    report.append("*Lower values indicate more consistent recommendations*")
    
    cv = all_weights.groupby("model")[weight_features].agg(
        lambda x: x.std() / x.mean() * 100  # CV as percentage
    )
    
    # Rename columns for display
    cv_display = cv.copy()
    cv_display.columns = [feature_display_names[col] for col in cv.columns]
    
    report.append("```")
    report.append(cv_display.to_string())
    report.append("```")
    
    # Top 3 features by model
    report.append("\n## Top 3 Features by Model")
    
    for model in all_weights["model"].unique():
        report.append(f"\n### {model}")
        model_means = means.loc[model].sort_values(ascending=False)
        top3 = model_means.head(3)
        
        for feature, value in top3.items():
            report.append(f"- **{feature_display_names[feature]}**: {value:.2f}")
    
    # Consensus analysis
    report.append("\n## Consensus Analysis")
    
    # Get rankings from each model
    rankings = {}
    for model in all_weights["model"].unique():
        model_means = means.loc[model]
        rankings[model] = model_means.rank(ascending=False).to_dict()
    
    # Calculate average rank for each feature
    feature_ranks = {}
    for feature in weight_features:
        feature_ranks[feature] = sum(rankings[model][feature] for model in rankings) / len(rankings)
    
    # Sort by average rank
    sorted_features = sorted(feature_ranks.items(), key=lambda x: x[1])
    
    report.append("\n### Overall Feature Ranking (By Average Rank)")
    for feature, avg_rank in sorted_features:
        report.append(f"- **{feature_display_names[feature]}**: {avg_rank:.1f}")
    
    # Identify clear consensus (top 2 and bottom 2 features)
    report.append("\n### Consensus Features")
    report.append("**Top 2 Most Important Features:**")
    for feature, avg_rank in sorted_features[:2]:
        report.append(f"- {feature_display_names[feature]}")
    
    report.append("\n**2 Least Important Features:**")
    for feature, avg_rank in sorted_features[-2:]:
        report.append(f"- {feature_display_names[feature]}")
    
    # Model-specific unique insights
    report.append("\n## Model-Specific Insights")
    
    # Identify features where each model differs significantly from the others
    for model in all_weights["model"].unique():
        report.append(f"\n### {model} Distinctive Emphasis")
        
        other_models = [m for m in all_weights["model"].unique() if m != model]
        other_means = all_weights[all_weights["model"].isin(other_models)].groupby("model")[weight_features].mean().mean()
        model_means = means.loc[model]
        
        # Calculate percentage difference
        pct_diff = (model_means - other_means) / other_means * 100
        
        # Find features with substantial differences (more than 25%)
        significant_diffs = [(feature, d) for feature, d in pct_diff.items() if abs(d) > 25]
        significant_diffs.sort(key=lambda x: abs(x[1]), reverse=True)
        
        if significant_diffs:
            for feature, diff in significant_diffs:
                direction = "higher" if diff > 0 else "lower"
                report.append(f"- **{feature_display_names[feature]}**: {abs(diff):.1f}% {direction} than other models")
        else:
            report.append("- No features with significantly different emphasis")
    
    # Recommendations for consensus weights
    report.append("\n## Consensus Weight Recommendation")
    
    # Calculate consensus weights (weighted by number of samples per model)
    weights = {}
    for feature in weight_features:
        numerator = sum(means.loc[model, feature] * model_counts[model] for model in means.index)
        denominator = sum(model_counts[model] for model in means.index)
        weights[feature] = numerator / denominator
    
    report.append("Based on the combined analysis of all three models, the consensus weights are:")
    for feature, weight in sorted(weights.items(), key=lambda x: x[1], reverse=True):
        report.append(f"- **{feature_display_names[feature]}**: {weight:.2f}")
    
    # Save the report
    with open(f"{report_path}/three_model_analysis_report_reanalyzed.md", "w") as f:
        f.write("\n".join(report))
    
    logger.info(f"Comparative report generated: {report_path}/three_model_analysis_report_reanalyzed.md")
def line_plot_weights_by_run(all_weights, base_path="/data/ep924610/project_nb/paper_code/bulk_review"):
    """
    Line plot: x-axis is run_id (or run index), y-axis is weight, each category is a line.
    One plot per model.
    """
    import matplotlib.pyplot as plt
    import seaborn as sns
    import os

    vis_path = f"{base_path}/reanalysis/visualizations"
    os.makedirs(vis_path, exist_ok=True)

    # Feature names for plotting
    weight_features = [
        "Clinical_trials", "HPA", "Heatmap", "Vital", 
        "Non_vital", "Protein_vital", "Protein_non_vital"
    ]
    
    feature_display_names = {
        "Clinical_trials": "Translatability",
        "HPA": "Subcellular Location",
        "Heatmap": "Cancer Composition",
        "Vital": "Specificity - Vital - Bulk RNA",
        "Non_vital": "Specificity - Non Vital - Bulk RNA",
        "Protein_vital": "Specificity - Vital - Protein",
        "Protein_non_vital": "Specificity - Non Vital - Protein"
    }

    model_colors = {
        "GPT-4o": "#1d6657",
        "Claude-3.7": "#b34700",
        "Gemini-2.5-Pro": "#0b5394"
    }
    
    # Make a plot for each model
    for model in all_weights["model"].unique():
        df = all_weights[all_weights["model"] == model].copy()
        # Sort by run_id for logical x-axis
        df = df.sort_values("run_id")
        plt.figure(figsize=(16, 8))
        
        for feature in weight_features:
            plt.plot(
                df["run_id"],
                df[feature],
                label=feature_display_names[feature],
                #marker='o'
            )
        
        plt.title(f"Run-wise Weight Recommendations for {model}", fontsize=18)
        plt.xlabel("Run ID", fontsize=14)
        plt.ylabel("Weight", fontsize=14)
        plt.legend(title="Feature", fontsize=12)
        plt.tight_layout()
        plt.grid(alpha=0.3)
        plt.savefig(f"{vis_path}/lineplot_weights_by_run_{model.replace('.', '').replace(' ', '')}.svg", dpi=300)
        plt.close()
    
    # Optionally, a combined spaghetti plot for all models (with different color for each feature, lines for all runs)
    plt.figure(figsize=(30, 10))
    for feature in weight_features:
        for model in all_weights["model"].unique():
            df = all_weights[all_weights["model"] == model].copy()
            df = df.sort_values("run_id")
            plt.plot(
                df["run_id"],
                df[feature],
                label=f"{feature_display_names[feature]} ({model})",
                alpha=0.3
            )
    plt.title("Run-wise Weight Recommendations (All Models)", fontsize=20)
    plt.xlabel("Run ID", fontsize=14)
    plt.ylabel("Weight", fontsize=14)
    plt.legend(fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.grid(alpha=0.3)
    plt.savefig(f"{vis_path}/lineplot_weights_by_run_all_models.svg", dpi=300)
    plt.close()

    print(f"Line plots saved to {vis_path}")
    
def line_plot_weights_by_run_by_feature(
    all_weights, 
    base_path="/data/ep924610/project_nb/paper_code/bulk_review",
    window=15
):
    """
    For each weight feature, plot run_id (x) vs. weight (y) for all models,
    with smoothed (rolling mean) lines and rolling std corridor.
    """
    import matplotlib.pyplot as plt
    import os

    vis_path = f"{base_path}/reanalysis/visualizations"
    os.makedirs(vis_path, exist_ok=True)

    weight_features = [
        "Clinical_trials", "HPA", "Heatmap", "Vital", 
        "Non_vital", "Protein_vital", "Protein_non_vital"
    ]
    feature_display_names = {
        "Clinical_trials": "Translatability",
        "HPA": "Subcellular Location",
        "Heatmap": "Cancer Composition",
        "Vital": "Specificity - Vital - Bulk RNA",
        "Non_vital": "Specificity - Non Vital - Bulk RNA",
        "Protein_vital": "Specificity - Vital - Protein",
        "Protein_non_vital": "Specificity - Non Vital - Protein"
    }
    model_colors = {
        "GPT-4o": "#74aa9c",       # OpenAI green
        "Claude-3.7": "#ff6c00",   # Anthropic orange
        "Gemini-2.5-Pro": "#4285F4" # Google blue
    }

    for feature in weight_features:
        plt.figure(figsize=(16, 8))
        for model in all_weights["model"].unique():
            df = all_weights[all_weights["model"] == model].sort_values("run_id")
            x = df["run_id"].values
            y = df[feature].values
            # Rolling stats
            if len(y) > window:
                import pandas as pd
                y_series = pd.Series(y)
                y_smooth = y_series.rolling(window=window, center=True, min_periods=1).mean()
                y_std = y_series.rolling(window=window, center=True, min_periods=1).std()
            else:
                y_smooth = y
                y_std = None
            # Plot mean line
            plt.plot(
                x,
                y_smooth,
                label=model,
                color=model_colors[model],
                linewidth=2.5
            )
            # Corridor (mean ± std)
            if y_std is not None:
                plt.fill_between(
                    x,
                    y_smooth - y_std,
                    y_smooth + y_std,
                    color=model_colors[model],
                    alpha=0.18,
                    linewidth=0
                )
        plt.title(f"Run-wise Weight for {feature_display_names[feature]}", fontsize=18)
        plt.xlabel("Run ID", fontsize=14)
        plt.ylabel("Weight", fontsize=14)
        plt.legend(fontsize=12)
        plt.tight_layout()
        plt.grid(alpha=0.3)
        plt.savefig(f"{vis_path}/lineplot_runwise_{feature}_smoothed_std.svg", dpi=300)
        plt.close()
    print(f"Feature-wise run line plots (smoothed, with std corridor) saved to {vis_path}")

# Usage:
# line_plot_weights_by_run_by_feature(all_weights, base_path, window=7)

# Usage example:
# line_plot_weights_by_run_by_feature(all_weights, base_path, smooth=True, window=7)
def relative_importance_plot(all_weights, base_path="/data/ep924610/project_nb/paper_code/bulk_review"):
    import matplotlib.pyplot as plt
    import seaborn as sns
    import numpy as np
    import os

    vis_path = f"{base_path}/reanalysis/visualizations"
    os.makedirs(vis_path, exist_ok=True)

    weight_features = [
        "Clinical_trials", "HPA", "Heatmap", "Vital", 
        "Non_vital", "Protein_vital", "Protein_non_vital"
    ]
    model_order = ["GPT-4o", "Claude-3.7", "Gemini-2.5-Pro"]
    model_colors = {
        "GPT-4o": "#1d6657",
        "Claude-3.7": "#b34700",
        "Gemini-2.5-Pro": "#0b5394"
    }
    feature_display_names = {
        "Clinical_trials": "Translatability",
        "HPA": "Subcellular Location",
        "Heatmap": "Cancer Composition",
        "Vital": "Specificity - Vital - Bulk RNA",
        "Non_vital": "Specificity - Non Vital - Bulk RNA",
        "Protein_vital": "Specificity - Vital - Protein",
        "Protein_non_vital": "Specificity - Non Vital - Protein"
    }
    # Calculate means for each model and feature
    means = all_weights.groupby("model")[weight_features].mean().loc[model_order]
    # Convert to relative (%) importance
    percent_importance = means.div(means.sum(axis=1), axis=0) * 100

    # Plot
    plt.figure(figsize=(12, 7))
    bottom = np.zeros(len(model_order))
    for feat in weight_features:
        plt.bar(
            model_order,
            percent_importance[feat],
            label=feature_display_names[feat],
            bottom=bottom
        )
        bottom += percent_importance[feat].values

    plt.ylabel("Relative Importance (% of total weight)", fontsize=14)
    plt.xlabel("Reasoning Model", fontsize=14)
    plt.title("Relative Importance of Features by Model", fontsize=16)
    plt.xticks(fontsize=13)
    plt.yticks(fontsize=13)
    plt.legend(title="Feature", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)
    plt.tight_layout()
    plt.savefig(f"{vis_path}/relative_importance_stackedbar.svg", dpi=300)
    plt.close()
    print(f"Relative importance plot saved to {vis_path}/relative_importance_stackedbar.svg")

def relative_importance_stackedbar_error_top(all_weights, base_path="/data/ep924610/project_nb/paper_code/bulk_review"):
    import matplotlib.pyplot as plt
    import numpy as np
    import pandas as pd
    import os

    vis_path = f"{base_path}/reanalysis/visualizations"
    os.makedirs(vis_path, exist_ok=True)

    weight_features = [
        "Clinical_trials", "HPA", "Heatmap", "Vital", 
        "Non_vital", "Protein_vital", "Protein_non_vital"
    ]
    model_order = ["GPT-4o", "Claude-3.7", "Gemini-2.5-Pro"]
    feature_colors = {
        "Clinical_trials": "#264653",
        "HPA": "#2a9d8f",
        "Heatmap": "#e76f51",
        "Vital": "#f4a261",
        "Non_vital": "#e9c46a",
        "Protein_vital": "#a061b5",
        "Protein_non_vital": "#5e548e"
    }
    feature_display_names = {
        "Clinical_trials": "Translatability",
        "HPA": "Subcellular Location",
        "Heatmap": "Cancer Composition",
        "Vital": "Specificity - Vital - Bulk RNA",
        "Non_vital": "Specificity - Non Vital - Bulk RNA",
        "Protein_vital": "Specificity - Vital - Protein",
        "Protein_non_vital": "Specificity - Non Vital - Protein"
    }

    # Calculate relative importance per run
    rel_weights = []
    for model in model_order:
        df = all_weights[all_weights["model"] == model]
        feature_vals = df[weight_features].values
        totals = feature_vals.sum(axis=1).reshape(-1, 1)
        rel_importance = (feature_vals / totals) * 100
        for i, row in enumerate(rel_importance):
            for j, feat in enumerate(weight_features):
                rel_weights.append({
                    "model": model,
                    "feature": feat,
                    "rel_imp": row[j]
                })
    rel_weights_df = pd.DataFrame(rel_weights)

    # Compute means and stds for error bars
    mean_imp = rel_weights_df.groupby(["model", "feature"])["rel_imp"].mean().unstack()
    std_imp = rel_weights_df.groupby(["model", "feature"])["rel_imp"].std().unstack()

    # Plot: stacked bars, errorbars at top of each segment
    x = np.arange(len(model_order))
    fig, ax = plt.subplots(figsize=(13, 7))
    bottoms = np.zeros(len(model_order))
    for i, feat in enumerate(weight_features):
        means = mean_imp[feat].reindex(model_order).values
        stds = std_imp[feat].reindex(model_order).values
        bars = ax.bar(
            x, means, bottom=bottoms, color=feature_colors[feat], 
            label=feature_display_names[feat], edgecolor="black", linewidth=0.7
        )
        # Error bars at the top of the segment
        bar_tops = bottoms + means
        ax.errorbar(
            x, bar_tops, yerr=stds, fmt='none', ecolor='black', capsize=6, lw=1.3, alpha=0.8, zorder=5
        )
        bottoms = bar_tops

    ax.set_xticks(x)
    ax.set_xticklabels(model_order, fontsize=13)
    ax.set_ylabel("Relative Importance (% of total weight)", fontsize=14)
    ax.set_xlabel("Reasoning Model", fontsize=14)
    ax.set_title("Relative Importance of Features by Model (error bars at top of segment)", fontsize=16)
    ax.legend(title="Feature", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=11)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    plt.tight_layout()
    plt.savefig(f"{vis_path}/relative_importance_stackedbar_error_top.svg", dpi=300)
    plt.close()
    print(f"Relative importance stacked bar plot (error bars at segment tops) saved to {vis_path}/relative_importance_stackedbar_error_top.svg")
def plot_weights_by_temperature(all_weights, base_path="/data/ep924610/project_nb/paper_code/bulk_review"):
    import matplotlib.pyplot as plt
    import seaborn as sns
    import os

    vis_path = f"{base_path}/reanalysis/visualizations"
    os.makedirs(vis_path, exist_ok=True)

    weight_features = [
        "Clinical_trials", "HPA", "Heatmap", "Vital", 
        "Non_vital", "Protein_vital", "Protein_non_vital"
    ]
    feature_display_names = {
        "Clinical_trials": "Translatability",
        "HPA": "Subcellular Location",
        "Heatmap": "Cancer Composition",
        "Vital": "Specificity - Vital - Bulk RNA",
        "Non_vital": "Specificity - Non Vital - Bulk RNA",
        "Protein_vital": "Specificity - Vital - Protein",
        "Protein_non_vital": "Specificity - Non Vital - Protein"
    }

    # For each feature, make a boxplot/violinplot grouped by model and temperature
    for feature in weight_features:
        plt.figure(figsize=(13, 7))
        sns.violinplot(
            x="temperature",
            y=feature,
            hue="model",
            data=all_weights,
            inner="box",
            split=True,
            cut=0,
            palette="Set2"
        )
        plt.title(f"Effect of Temperature on {feature_display_names[feature]}")
        plt.xlabel("Temperature")
        plt.ylabel("Weight")
        plt.legend(title="Model")
        plt.tight_layout()
        plt.savefig(f"{vis_path}/violin_{feature}_by_temperature.svg", dpi=300)
        plt.close()
    print(f"Temperature-feature violin plots saved to {vis_path}")
def anova_by_temperature(all_weights):
    import pandas as pd
    import scipy.stats as stats

    results = []
    for model in all_weights['model'].unique():
        df_model = all_weights[all_weights['model'] == model]
        for feature in [
            "Clinical_trials", "HPA", "Heatmap", "Vital", 
            "Non_vital", "Protein_vital", "Protein_non_vital"
        ]:
            groups = [
                g[feature].values
                for _, g in df_model.groupby("temperature")
                if len(g) > 1
            ]
            if len(groups) > 1:
                f_stat, p_val = stats.f_oneway(*groups)
                results.append({
                    "Model": model,
                    "Feature": feature,
                    "F-statistic": f_stat,
                    "p-value": p_val,
                    "Significant": "Yes" if p_val < 0.05 else "No"
                })
    return pd.DataFrame(results)
def summary_table_by_temp(all_weights):
    # Group by model and temperature for means/stds
    summary = all_weights.groupby(["model", "temperature"])[
        [
            "Clinical_trials", "HPA", "Heatmap", "Vital", 
            "Non_vital", "Protein_vital", "Protein_non_vital"
        ]
    ].agg(["mean", "std"]).round(2)
    return summary
def plot_mean_std_by_temperature(
    all_weights, 
    base_path="/data/ep924610/project_nb/paper_code/bulk_review"
):
    import matplotlib.pyplot as plt
    import numpy as np
    import os

    # Directory to save plots
    vis_path = f"{base_path}/reanalysis/visualizations"
    os.makedirs(vis_path, exist_ok=True)

    # Feature and label setup
    weight_features = [
        "Clinical_trials", "HPA", "Heatmap", "Vital", 
        "Non_vital", "Protein_vital", "Protein_non_vital"
    ]
    feature_display_names = {
        "Clinical_trials": "Translatability",
        "HPA": "Subcellular Location",
        "Heatmap": "Cancer Composition",
        "Vital": "Specificity - Vital - Bulk RNA",
        "Non_vital": "Specificity - Non Vital - Bulk RNA",
        "Protein_vital": "Specificity - Vital - Protein",
        "Protein_non_vital": "Specificity - Non Vital - Protein"
    }
    model_colors = {
        "GPT-4o": "#1d6657",
        "Claude-3.7": "#b34700",
        "Gemini-2.5-Pro": "#0b5394"
    }

    # Plot for each feature
    for feature in weight_features:
        plt.figure(figsize=(10, 6))
        for model in all_weights['model'].unique():
            df = all_weights[all_weights['model'] == model]
            # Group by temperature
            grouped = df.groupby("temperature")[feature].agg(['mean', 'std']).reset_index()
            # Sort by temperature
            grouped = grouped.sort_values("temperature")
            # Plot mean line
            plt.plot(
                grouped["temperature"], grouped["mean"], 
                marker="o", label=model, color=model_colors.get(model, None), linewidth=2
            )
            # Plot std corridor
            plt.fill_between(
                grouped["temperature"],
                grouped["mean"] - grouped["std"],
                grouped["mean"] + grouped["std"],
                alpha=0.18,
                color=model_colors.get(model, None)
            )
        plt.xlabel("Temperature", fontsize=13)
        plt.ylabel("Mean Weight (± std)", fontsize=13)
        plt.title(f"Mean (± std) {feature_display_names[feature]} vs. Temperature", fontsize=15)
        plt.legend(title="Model")
        plt.grid(alpha=0.3)
        plt.tight_layout()
        plt.savefig(f"{vis_path}/mean_std_vs_temperature_{feature}.svg", dpi=300)
        plt.close()
    print(f"Mean ± std plots of feature weights vs temperature saved to {vis_path}")
def temperature_effect_stats(
    all_weights,
    method='fdr_bh',  # Use Benjamini-Hochberg FDR correction
    base_path="/data/ep924610/project_nb/paper_code/bulk_review"
):
    import pandas as pd
    import statsmodels.formula.api as smf
    from statsmodels.stats.multitest import multipletests
    import numpy as np
    import os

    # Features for testing
    weight_features = [
        "Clinical_trials", "HPA", "Heatmap", "Vital", 
        "Non_vital", "Protein_vital", "Protein_non_vital"
    ]

    results = []
    for model in all_weights["model"].unique():
        df_model = all_weights[all_weights["model"] == model].copy()
        for feature in weight_features:
            # Remove rows with missing for this feature or temperature
            df_feat = df_model[["temperature", feature]].dropna()
            try:
                # Fit OLS: treat temperature as continuous predictor
                model_fit = smf.ols(f"{feature} ~ temperature", data=df_feat).fit()
                # Get p-value for temperature coefficient
                pval = model_fit.pvalues.get('temperature', np.nan)
                coef = model_fit.params.get('temperature', np.nan)
                results.append({
                    "Model": model,
                    "Feature": feature,
                    "N": len(df_feat),
                    "coef_temperature": coef,
                    "raw_p": pval,
                })
            except Exception as e:
                results.append({
                    "Model": model,
                    "Feature": feature,
                    "N": len(df_feat),
                    "coef_temperature": np.nan,
                    "raw_p": np.nan,
                })

    # Convert to DataFrame
    results_df = pd.DataFrame(results)

    # Adjust p-values for multiple testing (across all model-feature tests)
    reject, adj_p, _, _ = multipletests(results_df['raw_p'].values, method=method)
    results_df['adj_p'] = adj_p
    results_df['Significant'] = np.where(results_df['adj_p'] < 0.05, "Yes", "No")

    # Reorder and pretty print
    out = results_df[[
        "Model", "Feature", "N", "coef_temperature", "raw_p", "adj_p", "Significant"
    ]]
    out = out.sort_values(['Model', 'Feature'])

    # Save to CSV
    out_path = os.path.join(base_path, "reanalysis", "temperature_effect_stats.csv")
    out.to_csv(out_path, index=False)
    print(f"Temperature effect statistics (linear model, FDR adjusted) saved to {out_path}")

    return out
if __name__ == "__main__":
    # Set base path for your data
    base_path = "/data/ep924610/project_nb/paper_code/bulk_review"
    
    # Set debug level for detailed logging
    logger.setLevel(logging.DEBUG)
    
    # Reanalyze all model results with consistent extraction
    all_weights, summary, top_combos = reanalyze_all_models(base_path)
    relative_importance_plot(all_weights, base_path)
    # Create visualizations comparing all three models
    create_three_model_visualizations(all_weights, summary, base_path)
    relative_importance_stackedbar_error_top(all_weights, base_path)
    line_plot_weights_by_run(all_weights, base_path)
    line_plot_weights_by_run_by_feature(all_weights, base_path, window=10)
    # Generate comparative report
    generate_comparative_report(all_weights, summary, top_combos, base_path)
    plot_weights_by_temperature(all_weights, base_path)
    anova_results = anova_by_temperature(all_weights)
    summary = summary_table_by_temp(all_weights)
    print(summary)
    print(anova_results)
    plot_mean_std_by_temperature(all_weights, base_path)
    stats_df = temperature_effect_stats(all_weights, method='fdr_bh', base_path=base_path)
    print(stats_df)
    logger.info("Reanalysis complete! Check the updated visualizations and report.")
````
